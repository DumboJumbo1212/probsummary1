Status Graphs

```python
Status1 = dataframe1['STATUS'].value_counts()['resolved']
Status2 = dataframe1['STATUS'].value_counts()['reopened']
Status3 = dataframe1['STATUS'].value_counts()['open']
```

```python
status_counts = [Status1, Status2, Status3]
status_labels = ['Resolved', 'Reopened', 'Open']

plt.bar(status_labels, status_counts, color=['green', 'orange', 'blue'])
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Status Counts')
plt.show()

status_labels_2 = ['Reopened']
status_counts_2 = [Status2]

plt.figure(figsize=(6, 4))
plt.bar(status_labels_2, status_counts_2, color='orange')
plt.xlabel('Status')
plt.ylabel('Count')
plt.title('Status 2 (Reopened) Counts')
plt.show()
```

Affected Item Graphs

```python

Item_to_count = 'AFFECTED_ITEM'

Items_counts = dataframe1[Item_to_count].value_counts().to_dict()

Items_counts_df = pd.DataFrame(list(Items_counts.items()), columns=['Items', 'Count'])

Items_counts_df = Items_counts_df.sort_values(by='Count', ascending=False)

plt.figure(figsize=(15, 8))
plt.scatter(Items_counts_df['Items'], Items_counts_df['Count'])
plt.xlabel('Items')
plt.ylabel('Count')
plt.title('Items Counts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

City Of Issue

```python
column_to_count = 'RIL_CIRCLE'

State_counts = dataframe1[column_to_count].value_counts().to_dict()

State_counts_df = pd.DataFrame(list(State_counts.items()), columns=['State', 'Count'])

State_counts_df = State_counts_df.sort_values(by='Count', ascending=False)

plt.figure(figsize=(15, 8))
plt.scatter(State_counts_df['State'], State_counts_df['Count'])
plt.xlabel('State')
plt.ylabel('Count')
plt.title('State Counts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Define the columns to count
Item_to_count = 'AFFECTED_ITEM'
column_to_count = 'RIL_CIRCLE'

# Count the occurrences of each unique value in the columns
Items_counts = dataframe1[Item_to_count].value_counts().to_dict()
State_counts = dataframe1[column_to_count].value_counts().to_dict()

# Create DataFrames from the counts
Items_counts_df = pd.DataFrame(list(Items_counts.items()), columns=['Items', 'Count'])
State_counts_df = pd.DataFrame(list(State_counts.items()), columns=['State', 'Count'])

# Get sorted 'RIL_CIRCLE' values
sorted_states = State_counts_df['State'].sort_values()

# Split 'RIL_CIRCLE' values into three groups
num_groups = 3
group_size = len(sorted_states) // num_groups
split_indices = [i * group_size for i in range(1, num_groups)]
state_groups = np.array_split(sorted_states, split_indices)

# Plot each group in a separate graph
for i, state_group in enumerate(state_groups):
    # Filter DataFrame for the current group
    group_data = dataframe1[dataframe1[column_to_count].isin(state_group)]
    
    # Group by 'RIL_CIRCLE' and sum up counts of all affected items
    grouped_data = group_data.groupby(column_to_count)[Item_to_count].count().reset_index()
    
    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(grouped_data[column_to_count], grouped_data[Item_to_count],'bo')
    plt.xlabel('RIL_CIRCLE')
    plt.ylabel(f'Total {Item_to_count} Count')
    plt.title(f'Plot of Total {Item_to_count} Count against RIL_CIRCLE (Group {i+1})')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
